```python
#Checking if there exist duplicates or no
dupl_check=df.duplicated().tolist()
def dupl_checking(a): 
    b=0
    for i in a:
        
        if i==True:
            b+=1
    if b!=0:
        return "There's no duplicated value "
    else:
        return f"There is/are {b} duplicates "
dupl_checking(dupl_check)
```

```python
# checking if Unit cost is realistic or not
df.groupby(['_ProductID', 'ProcuredDate'])['Unit Cost'].nunique().reset_index()
```

```python
# here the code for finding the mean value for each product in the same day and converting it for further analyze 

# For deleting non needed spaces and replacing comma to dot 
df['Unit Cost'] = df['Unit Cost'].str.replace(',', '').str.strip()

# Converting data type to float
df['Unit Cost'] = df['Unit Cost'].astype(float)


df['Unit Cost'] = df.groupby(['_ProductID', 'ProcuredDate'])['Unit Cost'].transform('mean')

# Same for unit price 

# For deleting non needed spaces and replacing comma to dot 
df['Unit Price'] = df['Unit Price'].str.replace(',', '').str.strip()

# Converting data type to float
df['Unit Price'] = df['Unit Price'].astype(float)


df['Unit Price'] = df.groupby(['_ProductID', 'OrderDate', "Discount Applied"])['Unit Price'].transform('mean')
df



```

```python
df['RowNum'] = range(1, len(df)+1)
df.to_csv('processed_us_regional_sales.csv' )
```
